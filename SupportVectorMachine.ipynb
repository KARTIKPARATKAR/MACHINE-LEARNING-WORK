**Support Vector Machine-**

-->SVM is a supervised(Labelled Dataset) machine learning model.

-->SVM can be used for CLassification as well as Regression. But more predominantly SVM is used for Binary Classification problems like Male or Female and True or False.Binary classification problem is when we are predicting only two classes.

-->Hyperplan-It is a decision boundry that separates classes of the dataset

-->Support Vector-Nearest datapoints from hyperplane is called as Support Vectors.Support vectors are the closest data points to the Hyperplane in an SVM, defining the margin and influencing the hyperplane's position. Removing or changing them directly affects the decision boundary.Support Vector should be as far as from the Hyperplane so as to draw a clear Hyperplane.

-->Margin-It is distance between nearest support vectors and hyperplane.
