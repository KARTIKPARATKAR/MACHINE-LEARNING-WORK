Matplotlib library is used to plot the graphs of the data.

```python
#Importing matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy for data to plot the graphs
import numpy as np
```

```python
x= np.linspace(0,10,100)   #This will give 100 values which are equally spaced between 0 to 10
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#Plotting a data
plt.plot(x,y)  #Plotting graph of x vs y which wil show the sine wave
plt.show()
```

```python
plt.plot(x,z) #Plotting graph of x vs z which will show the graph of cosine wave
plt.show()
```

```python
plt.scatter(x,y)  #Scatter sin wave
```

```python
plt.stem(x,y)  #sA stem plot draws lines perpendicular to a baseline at each location locs from the baseline to heads, and places a marker there.
```

```python
plt.bar(x,y) #This will create a simple bar plot with one bar for each data point.
```

```python
#giving names to the x and y axis and giving name to the graph
plt.plot(x,y)
plt.xlabel('This is x-axis')         #X-axis naming
plt.ylabel('This is y-axis')         #Y-axis naming
plt.title('Name of the Graph')       #Graph naming
plt.show()
```

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.xlabel('X-values')
plt.ylabel('Y-values')
plt.title('Graph of parabola')
plt.show()
```

```python
#Now we want to plot the graph of this parabola with different colors
plt.plot(x,y,'r*')     #This will plot parabola with redcolor with * as mark
plt.xlabel('This is x values')
plt.ylabel('This is y values')
plt.title('Parabola with red color and * mark')
plt.show()
```

```python
x=np.linspace(-20,20,100)  #This will store the 100 values between -20 to 20 which are equally spaced
y=x**3                     #This is a definition of graph we want to draw
plt.plot(x,y,'g+')         #Plotting graph of x**3 with green color with sign +
plt.xlabel('This is x-values of x**3')       #Giving x axis name
plt.ylabel('This is y-values of x**3')       #Giving y axis name
plt.title('Graph of x**3')                   #Giving name to the graph
plt.show()
```

Now drawing multiple functions on single graph

```python
x=np.linspace(-5,5,20)
plt.plot(x,x**2,'r+')
plt.plot(x,x**3,'g+')
plt.title('Combining graph of x**2(Red) and x**4(Yellow)')
plt.xlabel('X-axis values')
plt.ylabel('Y-axis values')
plt.show()
```

Now plotting bar graphs to visualize data

```python
language=['Hindi','Marathi','Tamil','Telugu','Bangali']
Noofpeople=[20,100,40,85,60]
plt.bar(sorted(language),sorted(Noofpeople))  # Here we have used sorted so as to plot the bar in ascending order
plt.xlabel('Language')
plt.ylabel('No-of-people')
plt.title('Bar Plot of language vs No-og-people')
plt.show()

```

Plotting pie chart of the data.

```python
language=['Hindi','Marathi','Tamil','Telugu','Bangali']
Noofpeople=[20,100,40,85,60]
plt.pie(Noofpeople,labels=language,autopct='%0.2f%%')
plt.title('Pie plot for given data:')
plt.plot()
```

Scatter plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
plt.scatter(x,y)
plt.scatter(x,z)
plt.xlabel('X-data')
plt.xlabel('Y-data')
plt.title('Scatter plot of sine and cosine wave')
plt.plot()
```

Building 3-d scatter plot

```python
fig=plt.figure()
ax=plt.axes(projection='3d')  #Projection="3D" will create a 3D plot.
z=30*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='hsv')
plt.show()
```
