**Principle Component Analysis(PCA)-**

PCA is used for dimentionality reduction in machine learning.Whenever we are given high dimensional data and we want convert that data into lower dimentional data we use PCA.

1.jpg

**PCA** is well known unsupervised dimentionality reduction technique that constructs relevent features/variables through linear(PCA) or non-linear(kernel PCA) combinations of the original varibles(features).

**Now we will apply Principle Component Analysis on small data as follows:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
x = np.random.rand(100,5) #Creates a random dataset with 100 rows and 5 columns
```

```python
print(x) #Here note that we have 5 columns means we have 5 dimensions
#and these  dimensions we have to reduce to 2 so as to better training
#of any machine learning model on this dataset.
```

```python
scaler =StandardScaler() #Defining scaler function
```

```python
x_scaled = scaler.fit_transform(x) #This will standardized the data with zero mean and variance as 1
```

```python
print(x_scaled)
```

```python
pca = PCA(n_components=2) #Here we are defining the feature reduction to 2 from
x_pca = pca.fit_transform(x_scaled) # It computes PCA and transforms x_scaled into a lower-dimensional representation
```

```python
# Explained Variance
#Explain variance ratio tells how much information(variance) is captured by each principle
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
# Scatter Plot of Transformed Data
plt.scatter(x_pca[:, 0], x_pca[:, 1], c='blue', alpha=0.5)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Projection")
plt.show()
```

**So after executing this code, we had 100 samples and 5 features and now we have 100 samples and 2 principle components.**
