Categorical data is mostly in string and machine learning algorith needs numerical data.So we eill convert this string data into numerical data with the help of encoding technic called ordinal encoding and one-hot-encoding.

```python
import pandas as pd
import numpy as np
```

```python
df = df.iloc[:, 2:5]
#Here selecting 3rd,4th and 5th column in dataset

```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
#Here we are telling that give less value to poor, then average and then give highest value to good.
#Similarly give values in increasing order to School,UG and PG
```

```python
oe.fit(x_train) #Fitting on train data
```

```python
x_train=oe.transform(x_train)
x_test=oe.transform(x_test)
```

```python
x_train=pd.DataFrame(x_train)
```
