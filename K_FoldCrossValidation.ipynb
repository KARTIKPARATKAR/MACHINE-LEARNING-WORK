**K-Fold Cross Validation-**

Generally,We use train-test-split function to divide the dataset in training and testing so as to get the accuracy score.This is not the efficient way of evaluating the machine learning model. Instead of this, we use K-Fold-Cross-Validation.

In K-Fiold cross validation , we split the datset into "K" number of folds(subsets).One chunk of data is used as test data for evaluation and the remaining part of the data is used for training the model.Each time , a different chunk will be used as the test data.


1.jpg


In above example, we splitted the data into  chunks by defining the value of K as 5.There are 5 iterations we have. So while evaluating the model on each iteration,each time different chunks are used for training and different chunk is used for testing. And for every iteration , we will define a different instance of a model so that the model should be unaware of the dataset that they are training and testing.

**How K-Fold-Cross-Validation Used FOr Model Selection-**

Lets cosider we have a Regression problem and we have to select a Regression algorithm(SVM or Linear Regression) that works better on a given dataset. What we will do is , apply K-Fold-Cross-Validation on both the SVM and Linear Regression algorithm. If we define K=5 then we will get 5 different accuracy score for SVM and 5 different accuracy score for Linear Regression. Then we will take average of those 5 accuracy score for both algorithms. The one with higher average value of accuracy will be selected as final model.

**Advantage of K-Fold-Cross-Validation-**

- Better alternative for train-test-split when the dataset is small

- It is better for multiclass classification problem.

- It is more reliable as we are finding accuracy score on different chuncks of data with mutiple iterations.
