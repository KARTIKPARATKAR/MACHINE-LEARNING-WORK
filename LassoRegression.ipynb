**Lasso Regression-[Least Absolute Shrinkage and Selection Operator]**

It is a supervised regression machine learning algorithm. Regression means that model is predicting a numerical value like House Price,Blood Sugar Level and salary of a  employee etc.It uses (L1 Regularization) for avoiding Overfitting.
Lasso Regression is built on top of Linear Regression.

**Regularization-**

It is used to reduce the overfitting of the model by adding a penalty term () to the model.Lasso Regularization L1 regularization technique.

Consider the following Multiple-Linear-Regression problem where we are having 3-dependant variables w1,w2 and w3 .

1.jpg


As we have multiple dependant variables in Multiple-Linear-Regression , there is a high possibility of overfitting of the model. So using regularization technique , we will add penalty term which will remove the dependant variables which is not much contributing towards getting the output. As we are shrinking the number of dependant variables, so we got the name of this regularization as (LASSO-Least Absolute Shrinkage and Selection operator).

**Gradient Descent For Lasso Regression-**

Gradient Descent is the optimization algorithm used to minimize a loss function by iteratively updating the model's parameters(weights and biases).


What we do when we apply Gradient Descent algorithm on any ML/DL model is we draw weight[x] VS cost fn()[y] graph which is a parabola openinig upwards. Our aim to apply GD algorithm is to find the global minimum value of cost function so as to get better prediction accuracy. We assign random weight value and move towards global minimum value of cost function(). As soon as we reach to global minimum value,we consider that parameter value(w) for model training because that weight(w) value is giving least cost fn() value.
