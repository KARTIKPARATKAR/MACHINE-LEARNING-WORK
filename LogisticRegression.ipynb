**Logistic Regression-**

1)Logistic Regression is a supervised learning model

2)Logistic Regression is a classification model. Ex-News is fake or not, True Or False , Pink or Green.

3)Logistic Regression is best for binary classification problem. When we have only two output features to predict like true or false  then it is known as Binary Classification Problem.

4)Logistic Regression used Sigmoid Function.

**Sigmoid Function-**

The mathematical function that maps any real-valued number into a range between 0 to 1 is known as Sigmoid Function. SIgmoid function is widely used in Logistic Regression and Neural Networks for probability estimation.


sig(z) = 1/(1+e**(-z))

where, z=w.x + b
w is weight , b is bias , x is input  , sig(z) is probability

If sig(z) value is > 0.5 then model predicts 1 (positive)

If sig(z) value is < 0.5 then model predicts 0 (negative)

If your z value is higher positive value then value of sig(z) will have value close to 1.

If your z value is higher negative value then value of sig(z) will have value close to 0.

**Advantages of Logistic Regression** -

1) Easy to implement

2) Performs well on data with linear relationship

3) Less prone to over-fitting for low dimentional dataset



**Disadvantages ofLogistic Regression** -

1) High dimentional dataset causes over-fitting

2) Difficult to capture complex relationship in a dataset

3) Sensitive to outliers

4) Needs a larger dataset
