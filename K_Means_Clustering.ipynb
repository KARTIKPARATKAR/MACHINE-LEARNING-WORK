**Clustering-**

It is an unsupervised machine learning algorithm wherein the available dataset is unlabelled. Unlabelled means we dont know the output class of that datapoint.Clustering is used to group similar data points into clusters without labelled data.The goal here in clustering is to find pattern and structures in the dataset.

**K-Means-Clustering-**

K-Means CLustering is a technique used to organize data into groups based on their similarity.Here K in K-Means-Clustering is the number of groups/clusters we want to classify out items into.

1.jpg

**Hard Clustering-**

Each data point belongs to a cluster completely or not.If we have 4data point and we have to cluster them into 2 clusters.So each data point will either belong to cluster 1 or cluster 2.

**Soft Clustering-**

Instead of assigning each data point into a separate cluster,a likelihood of that point being that cluster is evaluated.If we have 4 data points and we have to cluster into 2 clusters.So we will be evaluating a probability of a data point belonging to both clusters.

**Centroid Based CLustering-**

Centroid of a cluster is the arithmatic mean of all the points in the cluster.Centroid-Based-CLustering organizes the data into non-hierarchical clusters.

**Density Based Clustering-**

It connects contiguous areas of high example density into clusters. This allows for the discovery of any number of clusters of any shape.Outliers are not assigned to clusters.

**Distribution-Based Clustering-**

As distance from the distribution's center increases,the probability that a point belongs to the distribution decreases.
