**Loss Function-**

Loss Function measures how far an estimated value is from its true value.We need to have minimum loss value. We can-not use the mean-squared-error loss function in Logistic Regression as it may lead us to multiple local minima. Therefore we will have to use Binary-Cross-Entropy loss Function (or) Log Loss function.


### Formula of BCE Loss
The Binary Cross-Entropy (BCE) Loss function is defined as:

BCE loss(y,y^) = -[(y*log(yi^)+(1-y)*log(1-y^)]

where:
- **n** = total number of samples  
- **y_i** = actual label (0 or 1)  
- **y_i'** = predicted probability (output of the sigmoid function)  
- **log** = natural logarithm  

The BCE loss is used for binary classification problems, ensuring that the model learns to predict probabilities close to 0 or 1.
