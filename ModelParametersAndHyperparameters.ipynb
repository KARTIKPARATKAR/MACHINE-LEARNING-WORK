Types of Parameters:

***Model Parameters:*** These are the parameters of the model that can be determined by training with training data. These can be considered as internal parameters.

--Weights

--Bias

If our model equation is:

y=wx+b , Here you can considers w as weight and b as bias. According to our data with which we are training the machine learning model, we will get the different values of w and b.


***Hyperparameters:*** These are parameters whose values control the learning process.These are adjustable parameters used to obtain an optimal model.These are external parameters.

--- Learning Rate(Number of change we are imparting to Parameter values i.e. w and b values)

--- Number of Epochs(No of times we train the model with perticular data)



***Weights-***

Weight decides how much influence the input will have on the output.
Number of weights in training a model is equal to the number of features available in dataset.

If feature has negative impact on output,then the value of weight for that feature will be negative.Eg-No of backlogs has negative impact on getting a job so here No of backlogs feature will have negative value of weight.


We may have mutiple features then equation of model will look like this:

y=w1x1 + w2x2 + w3x3 +b ;
here w1,w2,w3 are 3 different feature values and x1,x2,x3 are 3 different features, y is output and b is bias.

***Bias-***

Bias is the offset values to the model. Bias is used to shift the model in a particular direction. It is similar to a Y-intercept. 'b' is equal to 'y' when all the feature values are zero.If we use different dataset then our model equation will be different and as a result our weight and bias value will also be different.

***Hyperparameters-***

We will try to fit the model with data again and again so as to get the best weight and bias value to obtain a more accurate result.


***Learning Rate-***

It controls how much the model updates its weights during training. It determines the step size for moving toward a minimum of a loss function.

***No of Epochs-***

It represents the number of times the model iterates over the entire dataset. We have to give appropriate value of no of epoch otherwise our model can be underfit or overfit.

For determining the appropriate learning rate value and number of epochs value,we use Gradient Descent technique.
