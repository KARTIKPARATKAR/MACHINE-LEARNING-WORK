Seaborn is a Python data visualization library build on top of Matplotlib and closely integrated with Pandas.
We have some inbuild datasets in seaborn library that can be used to visualize and practice.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

Scatterplot in Seaborn

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
```

Lineplot in seaborn

```python
sns.lineplot(x='total_bill',y='tip',data=tips)
plt.xlabel("Total bills")
plt.ylabel("Tips")
plt.title("Plotting lineplot in seaborn")
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='size',hue='day',col='sex',size='size')
plt.xlabel("total_bill")
plt.ylabel("size")
plt.title("Plotting relplot in seaborn")
plt.show()
```

```python
df=tips
```

```python
df.head()
```

```python
sns.boxplot(x=df['tip'],data=df)
plt.show()
```

```python
sns.boxplot(x="sex",y="tip",data=df,hue='smoker')
plt.show()
```
