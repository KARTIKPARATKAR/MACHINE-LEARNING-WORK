**Accuracy Score**

Accuracy score is for CLassification problem. ,Accuracy Score is the ratio of number of correct predictions to the total number of input data points.
Let we have correct predictions =128
Total number of data points = 150

Accuracy score =(128/150)*100 = 85.3 %


We have a readmade function to calculate the accuracy score in machine learning in sklearn.metrices and you can use it by:

from sklearn.metrices import accuracy_score


**Mean Squared Error**

Mean squared Error is for Regression problem. Mean squared error measures the average of the squares of the errors,that is,the average squared difference between the estimated values and the actual value.

MSE = (1/n)*[sum of {(ypredic -yactual)**2}]
Where n is no of datapoints.

Acutal value is 140mg/dl

Predicted value is 160mg/dl

MSE = (1/1)(160-140)**2 = 20**2 = 400
So the mean squared error for 1 dataoint is 400.

We are desired to have a MSE value in regression problem to be 0. But this is not possible in most of the cases. Less the MSE value, better the prediction will be by that machine learning model.

Use the following code to import mean-squared-error in machine learing:
from sklearn.metrices import mean_squared_error
