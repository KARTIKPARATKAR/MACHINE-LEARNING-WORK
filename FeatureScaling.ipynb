Feature Scaling is done just before feeding the data to machine learning model.
Feature scaling is the easiest step in feature engineering.

Feature Scaling is a technique to standardize the independant features present in the data in a fixed range.

Feature Scaling has 2 types:Standardization and Normalization

When we do standardization of any column,then that column will have mean=0 and standard-deviation=1.

*Why we do standardization?
---->If we have two columns let age and salary to predict purchase or not. Here both columns age and salary is equally important to predict whether person will buy or not. But when we feed this age and salary data to machine learning model which is based on calculation of distance,it gives less importance to age columns because age columns will have lessor values as compared to the salary column.So to make sure that machine learing model give equal importance to both the columns , we standardize the both columns to the perticular range.

It is recommended that to do train test split before normalization of the data

```python
from sklearn.model_selection import train_test_split #IMporting  train-test-split module
```

```python
# X=df.drop('Purchased',axis=1)  #Dropping a column name 'Purchased' and storing it in X
# y=df['Purchased']              #Selecting target variable as 'Purchased' and assign it to y
# X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
# #test_size=0.25 means 25% data is reserved for test and 75% data is reserved for train
# #Random state =0 means ensure the split is reproducible each time you run the code
# X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Apply label encoding to categorical columns
X_train['Gender'] = label_encoder.fit_transform(X_train['Gender'])
X_test['Gender'] = label_encoder.transform(X_test['Gender'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(X_train) #Fit the scaler to the train set , it will learn the parameters
```

```python
#Transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled #It is numpy array. But we want datafram. So we will convert it into dataframe in next step.
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
X_train_scaled  #Now it is a pandas dataframe
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)  #Here we conclude that scaling is done because mean is zero and standard deviation is 1
```

Using LogisticRegression which is supervised machine learning algorithm.

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()  #Creating object of LogisticRegression
lr_scaled=LogisticRegression()
```

```python
lr.fit(X_train,y_train)  #Traing the logisticregression model with unscaled values
lr_scaled.fit(X_train_scaled,y_train) #Training the logisticregression model with scaled values
```

```python
y_pred=lr.predict(X_test)  #Prediction based on X_test data
y_pred_scaled=lr_scaled.predict(X_test_scaled)   #Prediction based on X_test_scaled data
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled))
```
