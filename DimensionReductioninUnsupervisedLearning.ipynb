**Dimension Reduction-**

Dimension Reduction is an Unsupervised machine learning technique which helps us to reduce large the number of feature columns to smaller number of feature columns.

If you have 500 feature column and you want to reduce the feature column numbers then use Dimension Reduction techniques to do so.

SO here in below fig, We have 3 dimensional data and we want this data to be in 2 dimension. So we will use dimension reduction and get the data converted into 2 dimension.

1.jpg

**Why Dimension Reduction-**

- If model is trained on unnecessary feature columns then it will take more time for training.

- If model is trained on differene complex feature columns then model will also be complex.

- If model is complex then prediction time will also be more.


**When we use Dimension Reduction in Machine Learning?**

**In Supervised Machine Learning-**

- To avoid overfitting

- To remove unnecessary information

- To reduce computational time.

**In Unsupervised Machine Learning-**

- To reduce dimensions to 2D/3D for easy interpretation.

- To reduce Noisy or redundant features to improve accuracy.

- To reduce complexity in clustering algorithm.

**In Reinforcement Learning-**

- To simplify agents decision making.

- To keep only useful state information

- To reduce computational complexity in deep reinforcement learning.
