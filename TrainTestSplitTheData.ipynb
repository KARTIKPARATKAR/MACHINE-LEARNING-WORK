```python
#Here we are splitting dataset into feature and target column. Features are those all columns except Outcome. And target column is Output
X=diabetesdataset.drop(columns='Outcome',axis=1)
Y=diabetesdataset['Outcome']
```

```python
print(X)  #Printing X which are the feature columns
```

```python
print(Y)  # Printing Y which is a target column
```

```python
#Now performing data standardization
scaler= StandardScaler()  #Data standardization is used to make all column values in a perticular standard range.
```

```python
scaler.fit(X)
```

```python
standardizeddata=scaler.transform(X)
```

```python
print(standardizeddata)
```

```python
X=standardizeddata
Y=diabetesdataset['Outcome']
```

Now splitting the data into training data and test data using train_test_split function.

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)  #Here test_size =0.2 means that 20% data will be used for testing the model and remaining 80 of data will be used for training the model.
```
