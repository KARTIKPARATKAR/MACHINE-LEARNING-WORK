**What is Dimentionality Reduction-**

It is a technique used to reduce the number of input variables present in a dataset.

If we have 100s of input variables then we will have to reduce the no of input variables so as to best train the model on dataset. This process of reducing the input features is known as Dimentionality Reduction.


**Dimentionality Reduction are of 2-types**

1)Feature Selection--

It is a process where u automatically or manually select those features which contribute most to your prediction variable or output.


2)Feature Extraction--

It is a  process by which an initial set of the data is reduced by identifying key features of the data for machine learning.

**What is Confusion Matrix in Classification Models?-->**

Confusion Matrix is used for evaluating the performance of a classification model.

False Negative--> Patient has corona but model is predicting as patient dont have corona.

True Negative--> Patient dont have corona and model is predicting as patient dont have corona.
