**What is a Machine Learning Model-->**

ML Model is a function that tries to find the relationship between the features and the target variables.

It tries to find the pattern in the data,understand the data and trains on the data. Based on this trainig, a Machine Learning Model makes predictions and recognize patterns.

Some of the examples of ML Models are- Logistic Regression,Support Vector Machine, K-Means Clustering etc.

**What are the different types of Machine Learning-->**

- Supervised Learning which includes classification and regression problems(Labelled Data used to train the model)
- Unsupervised Learning which includes clustering and association problems(Unlabelled Data is used to train the model)
- Re-Inforcement Learning-(It is an area of Machine Learning concerned with how intelligent agents take actions in an environment to maximize its rewards).Reinforcement learining is mostly used in game-playing-systems.

 In reinforcement learning we have 4 points-

 a)Environment(Chess Board)

 b)Agent (Computer on which we are playing this chess game)

 c)Action (Activities done to play the chess)

 d)Reward (If action is good then reward is positive and if action is not good then reward is negative )
