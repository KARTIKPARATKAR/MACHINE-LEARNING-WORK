Handaling Missing Values


```python
import pandas as pd
```

Use This Dataset: https://github.com/KARTIKPARATKAR/MACHINE-LEARNING-WORK/blob/main/Bengaluru_House_Data.csv

```python
df=pd.read_csv("/content/Bengaluru_House_Data.csv")
```

```python
df
```

```python
df.isnull().sum()  #Tells how many null values are there in each column
```

```python
df.shape  #13320 rows and 9 columns
```

Filling the null values by fillna

```python
df2=df.fillna(value=0)  #Here we are replacing null values with 0 in whole dataset and store it in df2
```

```python
df2.isnull().sum()  #As we have replaced null values with 0. This should give 0 in all the columns
```

```python
df3=df.fillna(value=5)  #Filling null values with 5 and storing it in df3
```

```python
df3 #Printing df3 where null values are replaced with 5
```

```python
  #Filling null values with previous value row-wise. It means value present above it
  dfn=df.ffill()
```

```python
dfn.isnull().sum()
```

```python
df6=df.fillna(method='bfill',axis=1)
```

```python
df6.isnull().sum()
```

```python
df6
```

```python
#Filling null value with the mean of column
df9=df.fillna(df.mean)
```

```python
df9.isnull().sum()
```

Dropna method

```python
df.head()
```

```python
df10=df.dropna() #This function will drop all the rows having null values and storing the dataset in df10
df10
```

```python
df11=df.dropna(axis=1) #This function will drop all the columns having null values and storing the dataset in df11
df11                   #Here axis=1 means column wise operation willbe done
```

```python
df11.isnull().sum()  #Here columns are dropped and only those columns are remaining where
                     #no null value is present
```

```python
df12=df.dropna()
```

```python
df12.head()
```

```python
df13=df.dropna(how='any') #This means that if any null value is present , then drop that column.
df13
```

```python
df14=df.dropna(how='all') #This means that if all the values in row is null then only drop that row
df14                      #After how='all', if i give axis=1 then it will apply for column
```

Replaceing null values


```python
import numpy as np
df15=df.replace(to_replace=np.nan,value=111111111)  #Replace null values with 11111111 this value
df15
```

```python
df16=df.replace(to_replace=3.0 , value=333333333333)  #Replacing value 3 with 33333333333333
df16
```

Using interpolate function for filling values

```python
df['balcony']=df['balcony'].interpolate(method='linear') #in balcony column , interpolate the balcony column and fill null value appropriately
df
```
