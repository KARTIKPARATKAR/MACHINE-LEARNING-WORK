**Parameters-**

**1)Model Parameters-**

These are the parameters of the model that can be determined by training with training data. These can be considered as internal Parameters.MOdel parameters are derived by the model by going through the data/by understanding the data.

EX-Weights and biases.

**2)Hyperparameters-**

These are parameters whose values control the learning process.These are adjustable parameters used to obtain an optimal model.External parameters.Hyperparameter is going to determine how your model is going to train.

Ex-Learning rate , number of epochs , n_estimator.


**Hyperparameter Tuning-**

It refers to the process of choosing the optimum set of hyperparameters for a Machine Learning model.This process is also called Hyperparameter Optimization.

1.jpg

When we talk about hyperparameter tunning , our aim is to find best hyperparameter and when we talk about model training , our aim is to find best model parameters.

**Hyperparameter Tunning Types-**

1)GridSearchCV-

We will consider all the datapoints present in dataset to check the accuracy of the model. We will check for what value of that perticular hyperparameter does machine learning model give better accuracy. GridSearchCV is not possible everytime when we work on the machine learning model because we can have a multiple combination of hyperparameter value.

2)RandomizedSearchCV-

We will consider a set of random datapoints present in dataset to check the accuracy of the model. We will check for what value of that perticular hyperparameter does machine learning model give better accuracy when set of random datapoints from avilable dataset is used.
