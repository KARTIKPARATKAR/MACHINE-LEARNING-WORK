**How Kernel Helps us to distinguish the datapoints-**

Lets consider dataset points as --

-3,-2,-1,0,1,2,3,4,5,6,7,8

Among these datapoints,-3,-2,-1,6,7,8 belongs to class A and 0,1,2,3,4,5 belongs to class B.

When you plot this dataset , it will be a straight line . We cannot distinguish this datapoints with the help of straight line no -matter-what. So what we will do is create a new feature by taking squares of eachdatapoint. We will get,

9,4,1,0,1,4,9,16,25,36,49,64

Now we will draw graph of this feature as (x,x^2).
We will get a parabola in which we can draw a straight line to distinguish the datapoints in class A and B.

**Types of SVM Kernel Function-**

Linear Kernel-->No of features in dataset is high as we already have a sufficient no of features available.

Polynomial Kernel--> No of features in a dataset is limited then use Polynomial Kernel in order to increase dimentional space.

Hard Margin--> We aim to classify all the points correctly,we assume the data is linearly separable by a hyperplane.

Soft Margin--> In real world problem,datasets are not linearly separable.In soft margine of SVM , we allow the model to misclassify some data points.We penalize the model for such misclassification.

There is no specific rule that when to use which Kernel function,whichever Kernel function is giving the best accuracy,we should use that Kernel function.

**Loss Function-**

Loss function measures how far an estimated value is from its true value.

It is helpful to determine which model performs are better & which parameters are better.If loss function value is less then model is performing better and if loss function value is high then model is not performing better.


Generally we use Squared Error loss to calculate the loss function value. But in case  of Support Vector Machine, We use **Hinge Loss**.

**Hinge Loss-**

Hinge Loss is one of the types of Loss Function,mainly used for maximum margin binary classification models.

Hinge Loss takes two details into consideration that a)Whether a datapoint is correctly classified or not?

b)Whether the margin distance is high or not?
