A dataset having imbalanced class destributioni.e.(1000 datapoint is of true and 100 datapoint is of false) is known as imbalanced dataset.
Dataset link is -https://drive.google.com/file/d/1WkcjUn5tVJDUtkN6d2_gztiT1FmIIVDP/view

```python
import numpy as np
import pandas as pd
```

```python
creditcarddata=pd.read_csv('/content/credit_data.csv')
```

```python
creditcarddata.head()  #1st five rows of the creditcarddata
```

```python
print(creditcarddata)  #There are 31 columns in a dataset with 9965 rows of data
```

```python
creditcarddata.tail() #Prints last  rows of the dataset
```

```python
#Destribution of the data class
creditcarddata['Class'].value_counts()  #Here we see that there are 9926 rows are of class 0 and only 38 rows are of 1.SO this is a imbalanced data.
#Here 0 represents legit transaction
#1 represents fraud transactions
```

```python
#Separating legit and fraudulant transaction
legit=creditcarddata[creditcarddata.Class==0]
fraud=creditcarddata[creditcarddata.Class==1]
#This will separate legit transaction and store it in variable legit
#And second line will separate fraud transaction and store it in fraud variable
```

```python
print(legit.shape)  #9926 rows in legit variable
print(fraud.shape)  #38 rows in fraud variable
```

Undersampling is a very important technique of handling imbalanced data.

```python
#Undersampling-
#Built a simple dataset containing similar distribution of legit and fraudulent transaction.
#NO of fraudulant transaction is 38
legitsample=legit.sample(n=40) #It will create 40 random values in legit dataframe and store it in legitsample dataframe
```

```python
print(legitsample.shape)
```

Concatinating two dataframes . 1st dataframe is legit and 2nd dataframe is legitsample

```python
newdataset=pd.concat([legitsample,fraud],axis=0)  #Axis = 0 means that we want to concatinate the data row wise
```

```python
newdataset.head() #First 5 data-row in newdataset
```

```python
newdataset.tail() #Last 5 datarow in newdatsset
```

```python
newdataset['Class'].value_counts()
```

Now here we have 40 data rows for 0 class and 38 data rows for 1 class. As a result we have got evenly destributed dataset which can be fed to the machine learning model.
