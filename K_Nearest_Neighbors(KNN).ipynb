**K-Nearest-Neighbors [KNN]-**

KNN is a supervised machine learning algorithm. KNN is used for both **Classification** as well as **Regression**. One of the advantage of using KNN is that KNN works better on Non-Linear data, which means that it works better on the data which sepread across.

**Consider a Classification problem to apply KNN-**

Here K is just a number which tells algorithm to consider the same no of datapoints as defined by K. If we define K value as (K=5) , then while classification of a perticular datapoint, KNN algorithm will consider the nearest 5 datapoints by calculating the Euclidian Distance(Generally) of new datapoint and the datapoints availble in datset.

Lets consider we have to classify this new datapoint in either Green or Blue.
Lets consider we have defined K value as 5. So what KNN algorithm will do is , it will find the nearest 5 datapoints from new datapoint by calculating Euclidian distance and among those 5 selected datapoints , it will see what is the majority category of those 5 points. If majority category Green then KNN algorithm will classify new datapoint in Green and if majority category is Blue then KNN algorithm will classify new datapoint in Blue.

**Consider a Regression Problem to apply KNN-**

Lets consider a problem of salary prediction based on number of years of experiance. Let we defined the value of k=5. So in regression problem what KNN algorithm will do is , it will calculate the nearest 5 datapoints by calculating Eucledian distance between new datapoint and datapoints in the dataset.

Now it will take the average salary of those 5 closest datapoints which we found out by calculating Eucledian distance.This average value of these 5 datapoints will be the salary of new-datapoint.

**Advantage of KNN-**

-Works well with smaller datasets with less number of features.

-Used for both classification as well as regression problem.

-Easy to implement for multi-class classification problems

-Different distance criteria can be used like Eucledian distance, Manhattan distance.

**Disadvantage of KNN-**

-Choosing optimum "K" value can make or break your model.

-Less efficient with high dimensional data.

-Doesnt perform well on imbalanced dataset.(Imbalnced dataseet means 1000 dataset for true and 20 dataset of false)

-It is very sensitive to outlayers

**Note-**

Manhatten distance calculation method is taken into consideration when we have dataset having higher dimensions.
