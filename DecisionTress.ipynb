**Advantages of Decision Tree-**

- It can be used for both Classification & Regression

- Easy to interpret

- No need for Normalization(Rescales data to a fixed range) or Scaling(Centers data around mean 0 with unit variance)

- Not sensitive to outliers

**Disadvantages of Decision Tree-**

- Overfitting issues

- Small changes in the data alter the tree structure causing instability

- Training time is relatively higher

**Entropy-**

Entropy measures the impurity or uncertainity in datsset.A high entropy value means the datset is more disordered while a low entropy value means the dataset is more pure.

1.jpg

**Information Gain-**

Information Gain tells us how useful a question or feature is for splitting data into groups.It measures how much the uncertainity decreases after the split.Low entropy leads to increased information gain and high entropy leads to low information gain.

**Gini Impurity-**

SPlit made in a decision tree is said to be pure if all the data points are accurately separated into different classes.Gini Impurity measures the likelihood that a randomly selected data point would be incorrectly classified by a specific node.
