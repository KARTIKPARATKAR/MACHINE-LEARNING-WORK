```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,20,100)  #Create a numpy array of 100 elements from 0 to 200 having evenly spaced
```

```python
x
```

```python
y=np.sin(x)
z=np.cos(x)
```

```python
print(y)
```

```python
print(z)
```

```python
plt.plot(x,y)  #Here y is sin values , so this plot is of sine
plt.xlabel('This is a X-axis')
plt.ylabel('This is a Y-axis')
plt.title("This plot is of SINE wave")
plt.show()
```

```python
plt.plot(x,z)  #Here z is cos values , so this plot is cosine values
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This plot is of COSINE wave")
plt.show()
```

```python
plt.scatter(x,y)  #Scatter sin wave
plt.xlabel("X-axis of scatter")
plt.ylabel("Y-axis of scatter")
plt.title("Sine wave of Scatter plot")
plt.show()
```

```python
plt.stem(x,y)
plt.xlabel("X-axis of stem plot")
plt.ylabel("Y-axis of stem plot")
plt.title("Stem plot of SINE wave")
plt.show()
```

```python
plt.bar(x,y)
plt.xlabel("X-axis of bar plot")
plt.ylabel("Y-axis of bar plot")
plt.title("Bar plot of SINE wave")
plt.show()
```

```python
p=np.linspace(-10,10,20)
print(p)
```

```python
q=p**2
```

```python
plt.stem(p,q)
plt.xlabel("X axis of Parabola")
plt.ylabel("Y axis of Parabola")
plt.title("Stem Parabola")
plt.show()

```

```python
plt.scatter(p,q)
plt.xlabel("X axis of Parabola")
plt.ylabel("Y axis of Parabola")
plt.title("Stem Parabola")
plt.show()
```

```python
plt.bar(p,q)
plt.xlabel("X axis of Parabola")
plt.ylabel("Y axis of Parabola")
plt.title("Bar Parabola")
plt.show()
```

```python
#Giving color to the plot
plt.plot(p,q,'r*')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Parabola with color")
plt.show()
```

```python
r=np.linspace(-50,50,100)
p=r**3
plt.plot(r,p,'r+')
plt.xlabel("X axis values of cubic plot")
plt.ylabel("Y axis values of cubic plot")
plt.title("Cubic plot with red color of +")
plt.show()
```

Drawing multiple functions on a single graph


```python
p=np.linspace(-10,10,100)
q=np.sin(p)
r=np.cos(p)
```

```python
plt.plot(p,q,'r*')
plt.plot(p,r,'g*')
plt.xlabel("X axis of multiple function")
plt.ylabel("Y axis of multiple function")
plt.title("Multiple function plot")
plt.show()
```

Plotting Bar Graph

```python
#Data for plotting
names=['A','B','C','D']
values=[20,60,19,30]
```

```python
plt.bar(names,values)
plt.xlabel("Names on X-axis")
plt.ylabel("Values on Y-axis")
plt.title("Bar Graph of Name and Values")
plt.show()
```

```python
plt.pie(values,labels=names,autopct='%1.2f%%')  #autopct='%1.2f%%'  Displays the percentage of each slice with 2 decimal places.
plt.title("Pie chart of Values and Names")
plt.show()
```

Scatter Plot
