```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()    #sets Seaborn's default aesthetic parameters to enhance the style and readability of plots.
```

Plotting the vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='g') #Here (0,0) means start vector from (0,0). Scale units "xy" means draw it on xy plane.
plt.xlim(-6,6) #Set x axis range from -6 to 6
plt.ylim(-6,6) #Set y axis range from -6 to 6
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='g') #Here (0,0) means start vector from (0,0). Scale units "xy" means draw it on xy plane.
plt.quiver(0,0,3,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-6,6) #Set x axis range from -6 to 6
plt.ylim(-6,6) #Set y axis range from -6 to 6
plt.show()
```

Addition of vector

```python
vector1=np.asarray([0,0, 2,3])
vector2=np.asarray([0,0,5,2])
sum = vector1 + vector2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g') #Here (0,0) means start vector from (0,0). Scale units "xy" means draw it on xy plane.
plt.quiver(0,0,5,2,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,7,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-6,10) #Set x axis range from -6 to 10
plt.ylim(-6,10) #Set y axis range from -6 to 10
plt.show()
```

Subtractio of Two Vectors

```python
vector1=np.asarray([0,0, 2,3])
vector2=np.asarray([0,0,5,2])
sum = vector1 - vector2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='g') #Here (0,0) means start vector from (0,0). Scale units "xy" means draw it on xy plane.
plt.quiver(0,0,5,2,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,1,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-6,10) #Set x axis range from -6 to 10
plt.ylim(-6,10) #Set y axis range from -6 to 10
plt.show()
```

Multiply Vector by scalar

```python
vec3=np.asarray([0,0,3,4])
vec4=2 * vec3
print(vec4)
```

```python
plt.quiver(0,0,6,8,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-6,10) #Set x axis range from -6 to 10
plt.ylim(-6,10) #Set y axis range from -6 to 10
plt.show()
```

```python
vec5=np.asarray([0,0,3,4])
vec6=-2 * vec5
print(vec6)
```

```python
plt.quiver(0,0,-6,-8,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-10,10) #Set x axis range from -10 to 10
plt.ylim(-10,10) #Set y axis range from -10 to 10
plt.show()
```
