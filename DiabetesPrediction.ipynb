**Problem Statement-**
We have a dataset of a patients which includes information of Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,BibetesPedigreeFunction,Age and Outcome whether the patient is diagnosed with Diabetes or not.
This is a classification problem wherein we have to apply classification ML algorithm to get output.

**Importing Dependencies-**

```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

**Data Collection and Analysis-**

```python
diabetes_dataset = pd.read_csv("/content/diabetes.csv")
```

```python
diabetes_dataset.head()
#Here in Outcome column, 1 reresents patient is diabetic and 0 represents patient is non-diabetic.
```

```python
diabetes_dataset.shape
#Rows are 768 and columns are 9
```

```python
#Getting the statistical measures of the dataset
diabetes_dataset.describe()
```

```python
#Check how many diabetic datapoints and non-diabetic datapoints
diabetes_dataset['Outcome'].value_counts()
```

0 --->  Non-Diabetic

1 ---> Diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
#This will give mean value of each column for diabetic datapoints and non-diabetic datapoints
```

```python
x = diabetes_dataset.drop(columns = 'Outcome' , axis = 1)
y = diabetes_dataset['Outcome']

#x will store only features by dropping the outcome column
#y will store the outcome column
```

```python
print(x)  #Data without outcome
```

```python
print(y)  #Data with only outcome
```

We observe that all the column values are in different range. It will be difficult for our ML-model to train on a dataset which not in a perticular range. To make a data in perticular range we will use standardscalar function which is a feature scaling technique.**StandardScalar function transform data to mean value of 0 and standard-deviation value of 1**

```python
scalar = StandardScaler()
```

```python
scalar.fit(x)
```

```python
standardized_data = scalar.transform(x)
```

```python
print(standardized_data)  #Here we observe that all the data is in a similar range
```

```python
x = standardized_data
y = diabetes_dataset['Outcome']
```

```python
print(x)
```

```python
print(y)
```

**Splitting data in training data and test data**

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.25 , stratify = y , random_state = 2)
#stratify = y means that our data will be splitted equally on the basis of diabetic datapoints and non-diabetic datapoints.
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

**Training the SVM Model-**


```python
svm_model = svm.SVC(kernel='sigmoid')
```

```python
svm_model.fit(x_train,y_train)
```

**Model Evaluation-**

```python
#Accuracy score on training data
x_train_accuracy = svm_model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_accuracy,y_train)
```

```python
print("Accuracy Score of Training data-",training_data_accuracy)
```

```python
#Accuracy Score on testing data
x_test_accuracy = svm_model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_accuracy,y_test)
```

```python
print("Accuracy Score on Testing data-",testing_data_accuracy)
```

**Building a predictive system-**

```python
input_data = (5,109,75,26,0,36,0.546,60)
#Changing input data as numpy array
input_data_as_numpy_array = np.asarray(input_data)

#Reshaping the array as we are predicting for one instance and not for all the datapoints available
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)


#When we trained the model,we standardized the data then trained the model so
#when we will predict the output by predictive system , we will need the standardized data only.
#So we will have to stndardized this input data.
std_data = scalar.transform(input_data_reshaped)

print(std_data)

prediction = svm_model.predict(std_data)
print(prediction)

if prediction[0] == 0 :
  print("The person is not diabetic")
else :
  print("The person is diabetic")
```
