**Linear Regression in Machine Learning**

Linear Regression is a supervised learning algorithm used for predicting continuous values. It establishes a relationship between independent variables (features) and a dependent variable (target) using a linear equation.

---

**ðŸ“Œ Types of Linear Regression**

**1. Simple Linear Regression** â†’ Involves a single independent variable.

**Equation:**  
\[ y = mx + c \]  
where:
- **y** = predicted value
- **m** = slope (coefficient)
- **x** = independent variable
- **c** = intercept

**2. Multiple Linear Regression** â†’ Involves multiple independent variables.

**Equation:**  
\[ y = b_0 + b_1x_1 + b_2x_2 + \dots + b_nx_n \]  
where:
- **y** = predicted value
- **b_0** = intercept
- **b_1, b_2, ..., b_n** = coefficients
- **x_1, x_2, ..., x_n** = independent variables
