Label Encoding is a technique used in machine learning and data analytics for converting categorical variables into numerical format


```python
#importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
#links for both the dataset are -
#https://drive.google.com/file/d/1j9aaJQ_kzFUEhSIYj2-jz-CC0zzPWjyN/view
#https://drive.google.com/file/d/1JrNVSa5HsfZ53y7-GQgHDiL1x7Ny7zi_/view
```

```python
#label encoding of breast cancer dataset i.e. dada.csv
#loading dataset
cancerdata=pd.read_csv('/content/data.csv')
```

```python
cancerdata.head()
```

```python
#COunting the different label
cancerdata['diagnosis'].value_counts()   #In diagnosis column,there are 357 labels are B and 212 labels are M. HEre we have calculated the no of M and B.
```

```python
#loading label cncoder function
labelencode=LabelEncoder()
```

```python
labels=labelencode.fit_transform(cancerdata.diagnosis)
```

```python
#appending the labels to dataframe
cancerdata['target']=labels   #This will include one column in dataset called target in which 0 represents B and 1 represents M.
```

```python
cancerdata.head()
```

```python
cancerdata['target'].value_counts()   #This means that in target column , we have 357 values for label 0 and 212 values for label 1
```

In above dataset i.e. data.csv, we had only 2 labels namely 0 and 1 i.e.(B and M)

In the below dataset i.e. iris_data.csv, we have 3 different labels

```python
irisdata=pd.read_csv('/content/iris_data.csv') #This will load the iris_data.csv to pandas dataframe
```

```python
irisdata.head()
```

```python
irisdata['Species'].value_counts()  #This will count the no of values for perticular label. Iris-setosa has 50 values, Iris_versicolor has 50 values and Iris_virginica has 50 values in "Species" column
```

```python
irisdata.head()   #This shows that target column is appended into the dataset called 'irisdata'
```
