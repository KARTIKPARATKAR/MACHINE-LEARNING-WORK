***Bias-Variance-Tradeoff***

Bias- It is the difference between the average prediction of our model and the correct value which we are trying to predict.


Variance- It is the amount that the estimate of the target function will change if different data was used.

Problem Statement: Identify an appropriate model to predict the height of a person WHen their weight is given.


When we are dealing wth Underfitted model,We have a High Bias and Low Variance and when we are dealing with overfitting model, We have low bias adn high variance.

A good model is one which does not have underfitting and overfitting. We call such models as optimized model. Bias-Variance tradeoff is a way of setting bias and variance values to get an optimized ML model.2.jpg



Techniques to have better Bias-Variance Tradeoff:

1) Good Model Selection

2) Regularization(Tries to reduce values of
   coefficents)   

3)Dimentionality Reduction -- If we have 1000s of features ,it makes your model more complex.We need to reduce this dimension.We will select features which are more important for getting the output.

4) Ensemble Methods -- It is something when we use 2-3 models to get better accuracy.
