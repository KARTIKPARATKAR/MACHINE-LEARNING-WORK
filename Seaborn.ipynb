Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics

```python
import seaborn as sns #importing seaborn library
import matplotlib as plt
import numpy as np
import pandas as pd
```

Seaborn has some inbuild datasets.

```python
# Total bill vs tips dataset
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#taking dataset tips. On x-axis take total_bill. On y-axis take tip. col='sex' means two different scatter plots will be created,one for sex-male and another for sex-female.
#hue='day' means that different days plot will be drawn by different color.size='size' means that more the no f size,biggere the size of circle
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',hue='day',size='size')
```

```python
#setting a theme for plots
sns.set_theme()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='day',hue='sex',size='size')   #here,we are plotting graph after setting a theme.
```

Now we will import another datset called 'IRIS'

```python
iris=sns.load_dataset('iris')   #iris is a dataset which has sepal length and width and petal length and width and species as a separate column.
```

```python
iris.head(5)
```

```python
sns.scatterplot(data=iris,x='sepal_length',y='petal_length',hue='species')    #on x-axis take sepal length . On y-axis take petal length. Graph color will bedifferentiated by its species.
```

```python
sns.scatterplot(data=iris,x='sepal_length',y='petal_width',hue='species')
```

Loading titanic dataset.

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()  #printing  rows of dataset titanic
```

```python
#count plot
sns.countplot(data=titanic,x='class')   #it will count the no of persons in 1st,2nd and 3rd class
```

```python
sns.countplot(x='survived',data=titanic)    #0 for not survived and 1 for survived.
```

Bar Chart

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```
